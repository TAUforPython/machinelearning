```python
import wfdb
import matplotlib.pyplot as plt
```

```python
#https://physionet.org/content/mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238/#files-panel

rec_path = '40689238'
ecg_wfdb_record = wfdb.rdrecord(rec_path, pn_dir='mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238/')
```

```python
y = wfdb.rdsamp(rec_path, channel_names='I', pn_dir='mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238/')
```
