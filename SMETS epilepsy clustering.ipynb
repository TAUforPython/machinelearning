```python
import numpy as np
import pandas as pd
import scipy.signal as signal
from scipy.stats import entropy
import mne
import pywt
from mne.preprocessing import ICA
import os
import glob
import matplotlib.pyplot as plt
import gc
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
import logging
from joblib import Parallel, delayed
from tslearn.metrics import dtw

# Настройка логирования
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Путь к данным
base_path = "C:/Users/chech/Downloads/masha-data/"
all_edf_files = glob.glob(os.path.join(base_path, "chb*/chb*_*.edf"))

# Разделение на нормальные файлы и файлы с приступами
normal_files = []
seizure_files = []
records_path = "C:/Users/chech/Downloads/masha-data/RECORDS-WITH-SEIZURES"

for edf_file in all_edf_files:
    seizure_file = edf_file + ".seizures"
    if os.path.exists(seizure_file):
        seizure_files.append(edf_file)
    else:
        normal_files.append(edf_file)

if not seizure_files and os.path.exists(records_path):
    with open(records_path, "r") as f:
        for line in f:
            file_name = line.strip()
            if file_name.endswith(".edf"):
                full_path = os.path.join(base_path, file_name)
                if os.path.exists(full_path):
                    seizure_files.append(full_path)
                    if full_path in normal_files:
                        normal_files.remove(full_path)

print(f"Нормальных файлов: {len(normal_files)}")
print(f"Файлов с приступами: {len(seizure_files)}")

# Ограничим количество файлов для анализа
seizure_files = seizure_files[:5]

# Функция загрузки данных
def load_eeg(file_path):
    try:
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
        return raw
    except Exception as e:
        logger.error(f"Ошибка при загрузке {file_path}: {e}")
        return None

# Функция фильтрации артефактов (обновленная)
def filter_artifacts(raw):
    if raw is None:
        return None
    # Полосовой фильтр с расширенным диапазоном
    raw.filter(l_freq=0.5, h_freq=70, method="iir")  # Обновлено до 70 Гц
    raw.notch_filter(freqs=50)

    # ICA с автоматическим выбором компонент
    ica = ICA(n_components=20, random_state=42)
    ica.fit(raw)
    # Выбираем компоненты с наибольшей дисперсией для исключения
    ica_sources = ica.get_sources(raw).get_data()
    variances = np.var(ica_sources, axis=1)
    exclude = np.argsort(variances)[-2:]  # Исключаем 2 компоненты с наибольшей дисперсией
    logger.info(f"Исключенные ICA-компоненты: {exclude}")
    raw = ica.apply(raw, exclude=exclude)
    return raw

# Функция нормализации данных
def normalize_data(raw):
    if raw is None:
        return None
    data = raw.get_data()
    mean = np.mean(data, axis=1, keepdims=True)
    std = np.std(data, axis=1, keepdims=True)
    normalized_data = (data - mean) / (std + 1e-10)
    raw._data = normalized_data
    return raw

# Функция сегментации данных
def segment_data(raw, window_size=1):
    if raw is None:
        return []
    sfreq = raw.info["sfreq"]
    window_samples = int(window_size * sfreq)
    segments = [
        raw[:, i : i + window_samples][0]
        for i in range(0, raw.n_times - window_samples + 1, window_samples)
    ]
    return segments

# Функция контроля качества (обновленная)
def quality_control(segments):
    if not segments:
        return []
    clean_segments = []
    for i, seg in enumerate(segments):
        max_abs = np.max(np.abs(seg))
        std = np.std(seg)
        threshold = 10 * std  # Уменьшенный порог
        if max_abs < threshold:
            clean_segments.append(seg)
        else:
            logger.info(f"Сегмент {i} отфильтрован: max_abs={max_abs:.2f}, threshold={threshold:.2f}")
    logger.info(f"Всего сегментов: {len(segments)}, Чистых сегментов: {len(clean_segments)}")
    return clean_segments

# Функция DWT преобразования
def dwt_transform(segment, level=6):
    if segment.size == 0:
        return np.array([])
    coeffs = pywt.wavedec(segment, "haar", level=level, axis=-1)
    all_coeffs = np.concatenate(coeffs, axis=-1)
    top_indices = np.argsort(np.abs(all_coeffs), axis=-1)[:, -47:]
    top_coeffs = np.take_along_axis(all_coeffs, top_indices, axis=-1)
    return top_coeffs

# Функция вычисления расстояний для SMETS
def compute_distances(coeffs1, coeffs2):
    if coeffs1.size == 0 or coeffs2.size == 0:
        return np.array([])
    distances = np.sqrt(
        np.sum((coeffs1[:, None, :] - coeffs2[None, :, :]) ** 2, axis=2)
    )
    return distances

# Функция парного соответствия
def pair_matching(distances):
    if distances.size == 0:
        return []
    d = []
    while distances.size > 0:
        min_dist = np.min(distances)
        idx = np.unravel_index(np.argmin(distances), distances.shape)
        d.append(min_dist)
        distances = np.delete(distances, idx[0], axis=0)
        distances = np.delete(distances, idx[1], axis=1)
    return d

# Функция p-нормы
def p_norm(d, p):
    if not d:
        return 0
    norm = np.sum(np.abs(d) ** p) ** (1 / p)
    return norm

# Функция энтропийного штрафа
def entropy_penalty(unmatched, coeffs1):
    EP = 0
    if unmatched and coeffs1.size > 0:
        for j in unmatched:
            H_j = entropy(np.histogram(j, bins=10, density=True)[0])
            min_d_j = min([np.min(np.sqrt(np.sum((j - c) ** 2))) for c in coeffs1])
            denom = sum(
                [entropy(np.histogram(c, bins=10, density=True)[0]) for c in coeffs1]
            )
            RE_j = min_d_j * H_j / denom if denom != 0 else 0
            EP += RE_j
    return EP

# Функция штрафа за различие в размерности
def dimensionality_penalty(m, n):
    return (n - m) / (n + m) if (m + n) > 0 else 0

# Функция SMETS
def SMETS(segment1, segment2):
    if segment1.size == 0 or segment2.size == 0:
        return 0, 0, 0
    coeffs1 = dwt_transform(segment1)
    coeffs2 = dwt_transform(segment2)
    if coeffs1.size == 0 or coeffs2.size == 0:
        return 0, 0, 0
    distances = compute_distances(coeffs1, coeffs2)
    d = pair_matching(distances)
    p = min(len(segment1), len(segment2))
    d_norm = p_norm(d, p)
    unmatched = []
    m, n = len(segment1), len(segment2)
    if m < n:
        unmatched = [coeffs2[i] for i in range(m, n)]
        EP = entropy_penalty(unmatched, coeffs1)
    elif n < m:
        unmatched = [coeffs1[i] for i in range(n, m)]
        EP = entropy_penalty(unmatched, coeffs2)
    else:
        EP = 0
    P = dimensionality_penalty(m, n)
    return np.sqrt((d_norm + EP) ** 2 + P**2), P, EP

# Функция чтения интервалов приступов
def get_seizure_intervals(seizure_file):
    intervals = []
    try:
        with open(seizure_file, 'r') as f:
            for line in f:
                if 'Start' in line:
                    start = float(line.split()[2])
                elif 'End' in line:
                    end = float(line.split()[2])
                    intervals.append((start, end))
    except Exception as e:
        logger.error(f"Ошибка чтения {seizure_file}: {e}")
    return intervals

# Функция маркировки сегментов
def label_segments(segments, seizure_intervals, window_size, sfreq):
    labels = []
    for i in range(len(segments)):
        start_time = i * window_size
        end_time = start_time + window_size
        is_seizure = any(start <= end_time and end >= start_time for start, end in seizure_intervals)
        labels.append(1 if is_seizure else 0)
    return labels

# Функция выбора подмножества сегментов
def select_segments(segments, labels, max_per_class=200):
    seizure_indices = [i for i, label in enumerate(labels) if label == 1]
    non_seizure_indices = [i for i, label in enumerate(labels) if label == 0]
    selected_seizure = seizure_indices[:max_per_class]
    selected_non_seizure = non_seizure_indices[:max_per_class]
    selected_indices = selected_seizure + selected_non_seizure
    selected_segments = [segments[i] for i in selected_indices]
    selected_labels = [labels[i] for i in selected_indices]
    logger.info(f"Выбрано приступных сегментов: {len(selected_seizure)}, нормальных: {len(selected_non_seizure)}")
    return selected_segments, selected_labels

# Функция вычисления евклидова расстояния
def euclidean_distance(segment1, segment2):
    return np.sqrt(np.sum((segment1 - segment2) ** 2))

# Функция вычисления DTW расстояния
def dtw_distance(segment1, segment2):
    # Для DTW используем усредненный канал (чтобы ускорить вычисления)
    seg1_mean = np.mean(segment1, axis=0)
    seg2_mean = np.mean(segment2, axis=0)
    return dtw(seg1_mean, seg2_mean)

# Вспомогательные функции для параллельного вычисления расстояний
def compute_pairwise_smets(i, j, segments):
    dist, _, _ = SMETS(segments[i], segments[j])
    return (i, j, dist)

def compute_pairwise_euclidean(i, j, segments):
    dist = euclidean_distance(segments[i], segments[j])
    return (i, j, dist)

def compute_pairwise_dtw(i, j, segments):
    dist = dtw_distance(segments[i], segments[j])
    return (i, j, dist)

# Функция вычисления матрицы расстояний с параллелизацией
def compute_distance_matrix(segments, metric='smets', n_jobs=16):
    n = len(segments)
    dist_matrix = np.zeros((n, n))
    pairs = [(i, j) for i in range(n) for j in range(i + 1, n)]

    if metric == 'smets':
        compute_func = compute_pairwise_smets
    elif metric == 'euclidean':
        compute_func = compute_pairwise_euclidean
    elif metric == 'dtw':
        compute_func = compute_pairwise_dtw
    else:
        raise ValueError(f"Неизвестная метрика: {metric}")

    results = Parallel(n_jobs=n_jobs, verbose=10)(delayed(compute_func)(i, j, segments) for i, j in pairs)
    for i, j, dist in results:
        dist_matrix[i, j] = dist
        dist_matrix[j, i] = dist
    return dist_matrix

# Функция кластеризации
def cluster_segments(dist_matrix, n_clusters=2):
    clustering = AgglomerativeClustering(
        n_clusters=n_clusters,
        metric='precomputed',
        linkage='average'
    )
    labels = clustering.fit_predict(dist_matrix)
    return labels

# Основная функция обработки данных
def process_chb_mit_data(seizure_files, window_size=1, max_segments_per_class=200):
    results = {}
    metrics = ['smets', 'euclidean', 'dtw']

    for metric in metrics:
        all_cluster_labels = []
        all_true_labels = []

        print(f"\n\n=== Обработка с метрикой {metric.upper()} ===")
        for file in seizure_files:
            print(f"\nОбработка файла: {file}")
            raw = load_eeg(file)
            if raw is None:
                continue
            raw = filter_artifacts(raw)
            raw = normalize_data(raw)
            segments = segment_data(raw, window_size=window_size)
            segments = quality_control(segments)
            if not segments:
                print("Чистых сегментов не найдено.")
                continue

            seizure_file = file + ".seizures"
            if os.path.exists(seizure_file):
                seizure_intervals = get_seizure_intervals(seizure_file)
                labels = label_segments(segments, seizure_intervals, window_size, raw.info['sfreq'])
                selected_segments, selected_labels = select_segments(segments, labels, max_per_class=max_segments_per_class)

                # Вычисление матрицы расстояний с параллелизацией
                dist_matrix = compute_distance_matrix(selected_segments, metric=metric, n_jobs=16)
                cluster_labels = cluster_segments(dist_matrix, n_clusters=2)

                print(f"Истинные метки (0 - норм, 1 - приступ): {selected_labels}")
                print(f"Метки кластеров: {cluster_labels}")

                all_cluster_labels.extend(cluster_labels)
                all_true_labels.extend(selected_labels)
            else:
                print("Файл аннотаций приступов не найден.")

            gc.collect()

        # Количественная оценка с помощью ARI
        if all_cluster_labels and all_true_labels:
            ari = adjusted_rand_score(all_true_labels, all_cluster_labels)
            print(f"Adjusted Rand Index для {metric.upper()}: {ari:.4f}")

            # Визуализация результатов
            plt.figure(figsize=(10, 6))
            plt.scatter(range(len(all_cluster_labels)), all_cluster_labels, c=all_true_labels, cmap='viridis')
            plt.title(f"Результаты кластеризации с {metric.upper()} (цвет - истинные метки)")
            plt.xlabel("Индекс сегмента")
            plt.ylabel("Метка кластера")
            plt.colorbar(label="Истинная метка (0 - норм, 1 - приступ)")
            plt.savefig(f"clustering_results_{metric}.png")
            print(f"График сохранен как 'clustering_results_{metric}.png'")

            # Гистограмма распределения кластеров
            plt.figure(figsize=(6, 4))
            plt.hist(all_cluster_labels, bins=2, rwidth=0.8)
            plt.title(f"Распределение кластеров ({metric.upper()})")
            plt.xlabel("Метка кластера")
            plt.ylabel("Количество сегментов")
            plt.savefig(f"cluster_distribution_{metric}.png")
            print(f"Гистограмма сохранена как 'cluster_distribution_{metric}.png'")

        results[metric] = (all_true_labels, all_cluster_labels)

    # Сравнение ARI для всех метрик
    print("\n=== Сравнение метрик ===")
    for metric in metrics:
        true_labels, cluster_labels = results[metric]
        ari = adjusted_rand_score(true_labels, cluster_labels)
        print(f"ARI для {metric.upper()}: {ari:.4f}")

# Запуск обработки
process_chb_mit_data(seizure_files)
```
